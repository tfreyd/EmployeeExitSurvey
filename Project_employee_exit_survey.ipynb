```python
#replace by NaN and True 
tafe_resignations['Contributing Factors. Job Dissatisfaction'].value_counts()
```

```python
#the magic function that will do everyting for us
def update_vals(val):
    if pd.isnull(val):
        return np.nan
    elif val == '-':
        return False
    else :
        return True 
```

```python
# apply to apply the function on both columns
col_tafe=['Contributing Factors. Dissatisfaction',
          'Contributing Factors. Job Dissatisfaction'
         ]

tafe_resignations['dissatisfied']=tafe_resignations[col_tafe].applymap(update_vals).any(axis=1, skipna=False)
tafe_resignations_up=tafe_resignations.copy()
tafe_resignations['dissatisfied'].value_counts(dropna=False)
```

```python
#same with dete survey dataset
col_dete=['job_dissatisfaction','dissatisfaction_with_the_department','physical_work_environment','lack_of_recognition','lack_of_job_security','work_location','employment_conditions','work_life_balance','workload']

dete_resignations['dissatisfied']=dete_resignations[col_dete].any(axis=1, skipna=False)
dete_resignations_up=dete_resignations.copy()
dete_resignations['dissatisfied'].value_counts()
```

### Clean the service column
employee's needs according to career stage instead of age is more effective.

We'll use the slightly modified definitions below:

* New: Less than 3 years at a company
* Experienced: 3-6 years at a company
* Established: 7-10 years at a company
* Veteran: 11 or more years at a company

```python
 combined_updated['institute_service'].value_counts(dropna=False)
```

```python
combined_updated['institute_service_up']=combined_updated['institute_service'].astype('str').str.extract(r'(\d+)')
```

```python
combined_updated['institute_service_up']=combined_updated['institute_service_up'].astype('float')
```

```python
combined_updated['institute_service_up'].value_counts()
```

```python
def career_stage(val):
    if pd.isnull(val):
        return np.nan
    elif val <3:
        return 'New'
    elif val <7:
        return 'Experienced'
    elif val < 11:
        return 'Established'
    else:
        return 'Veteran'
```

```python
combined_updated['service_cat'] = combined_updated['institute_service_up'].apply(career_stage)
```

```python
combined_updated['service_cat'].value_counts()
```
